## NUMPY MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,100,200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x")
plt.ylabel("This is my y")
plt.title("This is my data")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.scatter(x,y)
```

```python
x=['a','b','c','d','e']
y=np.random.rand(5)
plt.plot(x,y)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colours=np.random.rand(50)
plt.scatter(x,y,c=colours)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("This is my data")
plt.show()
```

```python
##For histogram data its shows the frequency of data
data=[1,2,4,5,6,7,8]
plt.hist(data)
plt.show()
```

```python
## For 3D data
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
## For scatterplot
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='smoker')
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='time',style='size')
```

```python
sns.catplot(x=tips.total_bill,y=tips.tip,data=tips)
```
